```python
import ee

ee.Authenticate()
```

```python
from google.colab import userdata
ee.Initialize(project=userdata.get('lab10_eeProject'))
```

```python
Map = geemap.Map(center=[20, 0], zoom=2)
Map.add_basemap('OpenStreetMap')

# Load the JRC Global River Flood Hazard Maps dataset
#https://developers.google.com/earth-engine/datasets/catalog/JRC_CEMS_GLOFAS_FloodHazard_v1#description
flood_hazard = ee.ImageCollection("JRC/CEMS_GLOFAS/FloodHazard/v1").mosaic()  #combine all images
depth = flood_hazard.select('depth')
depth_vis = {
    'min': 0.1,
    'max': 9495,
    'palette': ['#0000ff', '#00ffff', '#ffff00', '#ff7f00', '#ff0000', '#800000']
}
Map.addLayer(depth, depth_vis, 'Flood Inundation Depth')

Map
```

```python
output_file = 'index.html'
Map.to_html(filename=output_file)
```
